```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years=[1910 + i for i in range(10,110,10) ]
years
```

```python
vals=[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]
```

```python
ytick=[]
for i in range(0,100,1):
  ytick.append(i)
```

```python
sns.lineplot(x=years,y=vals,marker='o')
plt.xticks(years)

plt.xlabel("Years",fontsize=15)
plt.ylabel("Unemployment Rate",fontsize=15)
plt.title('Unemployment rate VS Years')
plt.fill_between(x=years,y1=vals,color='r')
"""
plt.fill_between(x=years,y1=vals,y2=100,color='g')"""
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 3, 4])
y = x*2
plt.plot(x, y)
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```
